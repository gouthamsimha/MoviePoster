```python
from moviepy.editor import VideoFileClip
from PIL import Image, ImageDraw
import os


input_video = '/Users/goutham/Downloads/gbh/gbh.mp4'
output_poster = '/Users/goutham/Downloads/gbh/gbh.png'
poster_height = 10492
poster_width = int(poster_height * 0.671)

# Check that input video file exists
if not os.path.isfile(input_video):
    raise FileNotFoundError("Input video file not found")

# Load the video file and scale it down if necessary
video = VideoFileClip(input_video)
if video.size[0] > poster_width or video.size[1] > poster_height:
    video = video.resize((poster_width, poster_height/video.size[0]*video.size[1]))

# Extract frames
frames = []
for t in range(0, int(video.duration), 1):
    frames.append(video.get_frame(t))

# Create the poster image
width, height = video.size
n_frames = len(frames)
cols = int((n_frames*height)/poster_height) # Calculate the number of columns based on the poster height
rows = int(n_frames/cols) + 1 # Calculate the number of rows based on the number of columns
h_gap, v_gap = 30, 30 # Set the horizontal and vertical gap between frames
poster = Image.new('RGB', (cols*(width+h_gap)-h_gap, rows*(height+v_gap)-v_gap), (0, 0, 0))
draw = ImageDraw.Draw(poster)

x, y = 0, 0
for i, frame in enumerate(frames):
    img = Image.fromarray(frame)
    poster.paste(img, (x, y))
    x += width + h_gap
    if (i+1) % cols == 0:
        x = 0
        y += height + v_gap

# Save the poster image
poster.save(output_poster)

# Load the poster image and calculate the gap sizes
poster_img = Image.open(output_poster)
left_gap = int(0.05 * poster_img.width)
right_gap = int(0.05 * poster_img.width)
top_gap = int(0.045 * poster_img.height)
bottom_gap = int(0.20 * poster_img.height)

# Create a new poster image with the gaps
width, height = poster.size
poster_with_gaps = Image.new('RGB', (width + left_gap + right_gap, height + top_gap + bottom_gap), (0, 0, 0))
poster_with_gaps.paste(poster, (left_gap, top_gap))

# Calculate the desired width based on the 0.671 aspect ratio
desired_width = int(poster_with_gaps.height * 0.671)

# Check if the current width is greater than the desired width
if poster_with_gaps.width > desired_width:
    # Resize the poster image to the desired width
    poster_with_gaps = poster_with_gaps.resize((desired_width, poster_with_gaps.height))

# Save the poster image with gaps and the correct aspect ratio
poster_with_gaps.save(output_poster)


print("Poster saved successfully!")
```
